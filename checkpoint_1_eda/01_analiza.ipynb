```python
import pandas as pd

url = r"../data/Road Accident Data.csv"

data = pd.read_csv(url, delimiter=',')

pd.set_option('display.max_columns', None)

print(data.head())
```

```python
cols_to_check = [
    'Urban_or_Rural_Area',
    'Local_Authority_(District)',
    'Police_Force'
]

check = (
    data.groupby(['Latitude', 'Longitude'])[cols_to_check]
    .nunique()
    .reset_index()
)

inconsistencies = check[
    (check[cols_to_check] > 1).any(axis=1)
]

print(f"Broj problematičnih lokacija: {len(inconsistencies)}")
print(inconsistencies)

details = (
    data.merge(
        inconsistencies[['Latitude', 'Longitude']],
        on=['Latitude', 'Longitude'],
        how='inner'
    )
    .sort_values(['Latitude', 'Longitude'])
)

print(details[
    [
        'Latitude',
        'Longitude',
        'Urban_or_Rural_Area',
        'Local_Authority_(District)',
        'Police_Force'
    ]
].drop_duplicates())
```
